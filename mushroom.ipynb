1. Target Variable Analysis (Class)

```python
df['class'].value_counts()
```

### Target Variable Distribution

The dataset contains two classes:
- Edible (e): 4208 samples
- Poisonous (p): 3916 samples

The edible class is the majority class, while the poisonous class is the minority class.  
However, the difference between the two classes is small, indicating that the dataset is nearly balanced.  
This balance is beneficial for classification tasks as it reduces bias toward a single class.


2.Cap Characteristics Analysis

```python
cap_features = ['cap-shape', 'cap-surface', 'cap-color']

for feature in cap_features:
    print(pd.crosstab(df[feature], df['class']))
    print("\n")
```

Cap features such as cap-shape, cap-surface, and cap-color show noticeable
differences between edible and poisonous mushrooms. Certain cap colors and surface
types appear more frequently in poisonous mushrooms, indicating that cap
characteristics are useful indicators of toxicity.


3. Bruises & Odor Analysis

```python
#(a) Bruises vs Class
pd.crosstab(df['bruises'], df['class'])
```

The presence or absence of bruises shows a relationship with mushroom edibility.
Mushrooms without bruises are more frequently poisonous, while bruising mushrooms
are more likely to be edible.


```python
#(b) Odor vs Class
pd.crosstab(df['odor'], df['class'])
```

Odor is one of the strongest indicators of mushroom toxicity. Certain odors such as
foul, fishy, pungent, and creosote are predominantly associated with poisonous
mushrooms, while edible mushrooms usually have mild or no odor.


4. Gill Features Analysis

```python
gill_features = ['gill-spacing', 'gill-size', 'gill-color']

for feature in gill_features:
    print(pd.crosstab(df[feature], df['class']))
    print("\n")
```

Gill features including gill-size, gill-spacing, and gill-color vary significantly
between edible and poisonous mushrooms. Certain gill colors and narrow gills are
more common in poisonous mushrooms, making gill characteristics strong predictors
of toxicity.


5. Stalk Characteristics Analysis

```python
stalk_features = [
    'stalk-shape',
    'stalk-root',
    'stalk-surface-above-ring',
    'stalk-surface-below-ring',
    'stalk-color-above-ring',
    'stalk-color-below-ring'
]

for feature in stalk_features:
    print(pd.crosstab(df[feature], df['class']))
    print("\n")
```

Stalk-related features such as stalk-root, stalk-shape, surface texture, and color
show distinct patterns across classes. Certain stalk roots and surface textures
occur more frequently in poisonous mushrooms, indicating their importance in
toxicity identification.


6. Veil & Ring Analysis

```python
veil_ring_features = ['veil-type', 'veil-color', 'ring-number', 'ring-type']

for feature in veil_ring_features:
    print(pd.crosstab(df[feature], df['class']))
    print("\n")
```

Veil and ring features show variation between edible and poisonous mushrooms.
Although veil-type is mostly constant, ring-number and ring-type differ across
classes, with some ring patterns appearing predominantly in poisonous mushrooms.


7. Spore, Population & Habitat Analysis

```python
pd.crosstab(df['habitat'], df['class'])
```

```python
pd.crosstab(df['population'], df['class'])
```

Habitat and population features show meaningful class-wise differences. Certain
habitats have a higher concentration of poisonous mushrooms, indicating that
environmental factors play an important role in mushroom toxicity.


8. Interaction & Combined Effects

```python
#(a) Odor × Gill Color
pd.crosstab([df['odor'], df['gill-color']], df['class'])
```

```python
#(b) Cap Color × Habitat
pd.crosstab([df['cap-color'], df['habitat']], df['class'])
```

Feature combinations such as odor with gill color and cap color with habitat show
stronger relationships with mushroom toxicity than individual features. Certain
combinations are highly indicative of poisonous mushrooms.


9. Feature Engineering Insights

```python
#(a) Toxic Odor Indicator
toxic_odors = ['f', 'y', 'c', 'p']
df['toxic_odor'] = df['odor'].isin(toxic_odors).astype(int)
```

```python
#(b) Cap Feature Combination
df['cap_combo'] = df['cap-shape'] + "_" + df['cap-color']
```

```python
#(c) Gill Complexity Score
df['gill_complexity'] = (
    df['gill-size'].map({'b': 1, 'n': 2}) +
    df['gill-spacing'].map({'c': 1, 'w': 2})
)
```

```python
#(d) High-Risk Habitat Indicator
high_risk_habitats = ['w', 'u', 'g']
df['high_risk_habitat'] = df['habitat'].isin(high_risk_habitats).astype(int)
```

New features were engineered to capture hidden patterns in the data, such as toxic
odor indicators, cap feature combinations, gill complexity scores, and high-risk
habitat indicators. These engineered features can improve the performance of
machine learning classification models.


1.Classify features into Qualitative and Quantitative
### Feature Classification: Qualitative vs Quantitative

The mushroom dataset contains only qualitative (categorical) features. All attributes,
including class, cap characteristics, gill features, stalk features, odor, habitat,
and population, represent categories rather than numerical measurements.

There are no quantitative (numerical) features present in the original dataset.
All features are encoded using categorical values.


2. Classify Features into 4 Levels of Measurement
🔹 Nominal Level

Categories with no natural order

Only labels / names

Mushroom features (MOST of them):

class (edible, poisonous)

cap-shape

cap-color

cap-surface

odor

gill-color

habitat

veil-type

veil-color

ring-type

stalk-color-above-ring

stalk-color-below-ring

📌 These features cannot be ordered meaningfully.

🔹 Ordinal Level

Categories with some order

Differences are not measurable

Mushroom features (few):

population (scattered < several < numerous)

gill-size (broad < narrow)

gill-spacing (close < crowded < distant)

ring-number (none < one < two)

📌 Order exists, but no numeric distance.

🔹 Interval Level

Numeric scale

Equal intervals

No true zero

❌ Not present in mushroom dataset

🔹 Ratio Level

Numeric scale

Equal intervals

True zero exists

❌ Not present in mushroom dataset

### Feature Classification Based on Levels of Measurement

The mushroom dataset features can be classified into four measurement levels.
Most features belong to the nominal level as they represent unordered categories
such as color, shape, odor, and habitat. A few features such as population,
gill-size, gill-spacing, and ring-number are ordinal in nature as they have a
natural order.

There are no interval or ratio level features in the mushroom dataset, as it
does not contain any numerical measurements.
